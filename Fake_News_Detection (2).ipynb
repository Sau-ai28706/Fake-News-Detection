```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import re
import joblib
import string
```

```python
fake = pd.read_csv('/content/drive/MyDrive/ML_Datasets/Fake.csv.zip')
true = pd.read_csv('/content/drive/MyDrive/ML_Datasets/True.csv.zip')
```

```python
fake.head()
```

```python
true.head()
```

```python
fake['class'] = 0
true['class'] = 1
```

```python
data = pd.concat([fake,true], axis = 0)
```

```python
data.sample(10)
```

```python
data = data.drop(['title','subject','date'], axis = 1)
```

```python
data.reset_index(inplace = True)
```

```python
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.sample(10)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
data["text"] = data["text"].apply(clean_text)
```

```python
x = data["text"]
y = data["class"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25,
                                                random_state = 42)
```

```python
vectorizer = TfidfVectorizer()
xtrain_vectorized = vectorizer.fit_transform(xtrain)
xtest_vectorized = vectorizer.transform(xtest)
```

```python
lr = LogisticRegression()
lr.fit(xtrain_vectorized, ytrain)
```

```python
prediction = lr.predict(xtest_vectorized)
lr.score(xtest_vectorized, ytest)
```

```python
print(classification_report(ytest, prediction))
```

```python
joblib.dump(vectorizer, 'vectorizer.pkl')
joblib.dump(lr, 'model.pkl')
```
